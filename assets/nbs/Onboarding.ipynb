## 🧠 Give the Machine Memory

Variables disappear after Notebooks *restart.* Pipulate uses the `wand` for memory. Let's set a target URL for our first analysis and save it to the local database.

```python
# Step 2: Define the target
TARGET_URL = "https://example.com"

wand.set("onboarding_job", "target_url", TARGET_URL)  # <-- wand.set()

wand.speak(
    f'The "target_url" key has been set to the value "{TARGET_URL}"\n'
    f'for the Workflow whose Job-ID is "onboarding_job". See how key-value memory works?'
)

wand.imperio()
```

> Note of how `wand.set()` ***always*** uses a **Job ID** (like "onboarding_job") when recording a value.

## 🔎 Inspect Wand Memory

If you use `wand.set()` then it remembers. But what does it remember, exactly? Run the next cell to peek inside the `wand`'s memory for this job.

```python
import json
from pipulate import wand

# We ask the wand to read the entire state of our 'onboarding_job'
current_state = wand.read("onboarding_job")

wand.speak("This is persistent memory. What the wand writes, the wand can read again.\n"
           "This caching ensures we can resume workflows and eventually port them into web apps.\n"
           "Now that we have memory, it's time to give the machine eyes. Get ready for browser automation.")

print("🧠 Pipulate's Current Memory State:\n")
print(json.dumps(current_state, indent=2))

wand.imperio()
```

## 👀 Give the Machine Eyes

We have a voice, and we have persistent memory. Now, we need optics. It's time for (local) browser automation!

**Run the next cell.** *Keep your hands off the mouse* and watch the machine work *(do not close the browser that pops up).* The 20-second delay of the browser just sitting there is how we prove to the site you are human.

```python
# Step 3: Execute the scrape

wand.speak(
    f"Initializing browser automation for {recovered_url}. "
    "Wait for the browser to close itself. This could take up to 30 seconds. \n"
    "Be patient — we are waiting out an invisible CAPTCHA to prove to the server \n"
    "that you are a carbon-based lifeform. Hands off the mouse!",
    delay=1.5,     # <-- Give the browser a second to pop up before talking
    wait=False     # <-- Non-blocking! The cell moves immediately to wand.scrape
)

print("We are waiting out an invisible CAPTCHA...")

result = await wand.scrape(
    url=recovered_url, 
    headless=False,
    override_cache=True
)

if result.get('success'):
    if result.get('cached'):
        wand.speak("Cache Hit! Using existing artifacts. If you want to see the browser pop up again, change override_cache to True.")
    else:
        wand.speak("Scrape Successful.")
else:
    wand.speak("I encountered an error during navigation.")
    print(f"Scrape Failed: {result.get('error')}")

wand.imperio()
```

> Note: The results of that scrape are saved locally for you to examine.

## 🤖 Select Your AI

Pipulate uses both local and cloud-model AIs. We set up a preferred choice for each. We use **[Ollama](https://ollama.com/)** to provide the local AI from the Host OS, which is not installed by Pipulate. So if you want local AI you must *[install it yourself](https://ollama.com/).* I suggest `gemma4:latest` or `qwen3.5:latest`. Ollama runs on your Host OS as a service and can be activated or deactivated at any time.

```python
# Define your AI hierarchy. The system will attempt to use the first available model in the list.
PREFERRED_LOCAL_AI = "gemma4:latest, qwen3.5:latest"
PREFERRED_CLOUD_AI = "gemini, claude, gpt"

wand.speak("Let's look for your models.")

ai_status = wand.negotiate_ai_models(
    preferred_local=PREFERRED_LOCAL_AI, 
    preferred_cloud=PREFERRED_CLOUD_AI
)

if isinstance(ai_status["local"], str):
    ACTIVE_MODEL = ai_status["local"]
else:
    ACTIVE_MODEL = "gemma4" 

wand.speak("Models negotiated. Checking the vault for cloud credentials...")

# The Gatekeeper: Forces the API key request up-front via a secure widget if missing.
# If it's already in your .env, it silently passes and calls wand.imperio()
sauce.ensure_cloud_credentials(ai_status["cloud"])

wand.imperio()
```

```python
# Step 4: The Local Assessment & The Excel Egress

# 1. Prepare the AI directives (Path and I/O logic pushed to sauce)
system_prompt, user_prompt = sauce.build_local_optics_prompt(recovered_url)

wand.speak(f"Channeling local intent through {ACTIVE_MODEL} to analyze the optics...")

# 2. Execute the local prompt (Kept front-and-center for the user to see)
ai_assessment = wand.prompt(
    prompt_text=user_prompt, 
    model_name=ACTIVE_MODEL, 
    system_prompt=system_prompt
)

# Stripped down display to keep the cell dependencies ultra-light
print(f"\n🤖 Chip O'Theseus ({ACTIVE_MODEL}):\n")
print(ai_assessment)
print("\n" + "-"*40 + "\n")

# 3. Package the deliverable
wand.speak("Structuring the optical artifacts into a client-ready Excel deliverable.")

button, xl_file = sauce.package_optics_to_excel("onboarding_job", recovered_url, ai_assessment)

display(button)
print(f"💾 Optics Baseline Generated: {xl_file.name}")

wand.speak(
    "Deliverable created. We have successfully translated raw browser automation "
    "into tangible intelligence. You may open the folder to inspect the result. \n"
    "But local AI is just the beginning. Run the next cell to package our optics into a "
    "JavaScript Gap analysis ready for a frontier Cloud model."
)

wand.imperio()
```
