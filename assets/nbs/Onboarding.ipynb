### Give the Machine Memory

Variables disappear after Notebooks *restart.* Pipulate uses the `wand` for memory. Let's set a target URL for our first analysis and save it to the local database.

```python
# Step 2: Define the target
TARGET_URL = "https://example.com"

wand.set("onboarding_job", "target_url", TARGET_URL)  # <-- wand.set()

wand.speak(f"Target saved to database: {TARGET_URL}")

wand.speak("Done step. Please run the next cell.")
```

**Restart the Kernel**. Press the `Esc` key, then type `0`, `0`. 

Normally, this would wipe the notebook's memory entirely. Run the next cell to see how the `wand` recovers your state. This mechanism is the foundation for turning these explicit, top-to-bottom scripts into robust web applications.

```python
from pipulate import wand  # <-- Accio Wand!

recovered_url = wand.get("onboarding_job", "target_url")  # <-- wand.get()

try:
    NAME
    wand.speak(f"Target is: {recovered_url}. But you didn't reset the kernel so I still remember your name is {NAME}.")    
except:
    wand.speak(f"State recovered. Target is: {recovered_url}. But now I forget your name.")

print("Done step. Run next cell.")
```

> Notice how the **wand** was *re-imported.* That's okay. Accio Wand as much as you like!

### Give the Machine Eyes

We have a voice, and we have persistent memory. Now, we need optics. It's time for (local) browser automation!

**Run the next cell.** *Keep your hands off the mouse* and watch the machine work *(do not close the browser that pops up).*

```python
# Step 3: Execute the scrape

wand.speak(f"Initializing browser automation for {recovered_url}.")

result = await wand.scrape(
    url=recovered_url, 
    take_screenshot=True, 
    headless=False,        # <-- headless=False means the browser window will pop up on your screen (if not cached).
    override_cache=False,  # <-- Set this to True if you want to force the browser to open and pull fresh data
    verbose=True
)

if result.get('success'):
    if result.get('cached'):
        wand.speak("Cache Hit! Using existing artifacts. If you want to see the browser pop up again, change override_cache to True.")
    else:
        wand.speak("✅ Fresh Scrape Successful.")
else:
    wand.speak("I encountered an error during navigation.")
    print(f"Scrape Failed: {result.get('error')}")

print("Done step. Run next cell.")
```

### The Transparent Memory

The browser opened, evaluated the page, and closed. But what did Pipulate actually remember? 

A cloud service hides your data in a black box. A sovereign machine shows you exactly what it knows. Run the next cell to peek inside the `wand`'s memory for this job.

### Pointers, Not Payloads (The Side Effects)

To prepare the web for an AI, we cannot simply extract text. We must translate the DOM into a format a Large Language Model can comprehend. Because you are running Pipulate locally, we can make this process visible. 

Notice that the memory state above does *not* contain the massive HTML file of the website, nor the image data for the screenshot. 

Storing massive files in a database makes systems slow and brittle. Instead, Pipulate stores the heavy data as standard files directly on your hard drive (in the `browser_cache/` folder), and the database merely acts as a pointer. 

Let's use the URL we saved to the `wand` earlier to locate those files.

```python
import json
from pipulate import wand

# We ask the wand to read the entire state of our 'onboarding_job'
current_state = wand.read("onboarding_job")

print("🧠 Pipulate's Current Memory State:\n")
print(json.dumps(current_state, indent=2))

print("Done Step.")
```
