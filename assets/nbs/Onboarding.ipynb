## ☁️ Connect the Cloud AI

While your Local AI is great for privacy and unlimited free usage, sometimes you need a **"Specialized Subcontractor"**—a frontier cloud model—for heavy lifting. 

To connect a Cloud AI reliably, you need an **API Key**, not just a consumer subscription. 
* **Subscriptions (The "Netflix Login")**: Your $20/mo ChatGPT Plus or Claude Pro account uses an "OAuth token". It's meant for humans who sleep. If you use tools that hijack this connection, it will eventually expire or get rate-limited while your automated workflow runs.
* **API Keys (The "Electric Bill")**: A permanent credential intended for a machine. It connects directly to the provider's developer platform. It acts like a metered utility connection and rarely expires.

**Where to get your API Key (Free Tiers):**
1. **Google (AI Studio) - 🏆 Recommended**: Offers a massive, daily-renewing free tier for models like `gemini-1.5-flash`. Just go to [Google AI Studio](https://aistudio.google.com/) with your Gmail account and generate an API key. 
2. **Groq**: Offers a very generous and incredibly fast free tier for open-source models. Visit [GroqCloud](https://console.groq.com/).
3. **Anthropic & OpenAI**: These are strictly business. They generally do not offer permanent free API tiers. You must add a credit card and pre-pay for usage.

```python
from pipulate import wand  # <-- Pipulate magic wand
from imports import onboard_sauce as sauce

# Define your AI hierarchy. The system will attempt to use the first available model in the list.
PREFERRED_CLOUD_AI = "gemini, claude, gpt"

wand.speak(
    'Now let\'s bring in the heavy machinery. While it\'s possible to use "consumer" OAuth web-logins \n'
    '[through Claude Code or Codex (if you know what you\'re doing)], getting a formal API Key[ — the metered] \n'
    "['Electric Bill' method —] is more reliable for making your cloud-consulting robots unstoppable. \n" 
    "I recommend getting a free API-key for Gemini from [Google AI Studio](https://aistudio.google.com/api-keys) to start. \n\n"
    "[(I'd use Claude or ChatGPT but Google's free-tier API-keys are so available and suitable here.)] \n\n"
)

# Normal execution
ACTIVE_CLOUD_MODEL, KEY_READY = wand.verify_cloud_ai(preferred_models=PREFERRED_CLOUD_AI)

# --- BEGIN DEBUG CONTROLS ---

# Force it to think the API key is missing for testing
# ACTIVE_CLOUD_MODEL, KEY_READY = wand.verify_cloud_ai(
#     preferred_models=PREFERRED_CLOUD_AI, 
#     simulate_state='missing_key'
# )

# --- END DEBUG CONTROLS ---

if KEY_READY:
    wand.imperio()
else:
    # Deliberate Side Quest. The widget will issue wand.imperio() once the key is saved.
    wand.imperio(side_quest="optional")
```
