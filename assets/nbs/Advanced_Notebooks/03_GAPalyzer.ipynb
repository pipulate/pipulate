```python
# 🤖 Expert Consulting: Interactive Prompt Generator (Ultimate Edition)

import pandas as pd
import re
import json
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- 1. Hydrate Missing Variables from Pipulate State ---
semrush_lookup = secretsauce._extract_registered_domain(keys.client_domain)
competitors_json = wand.get(job, 'competitors_list_json', '[]')
competitors = json.loads(competitors_json)

if not competitors:
    clean_lookup_key = semrush_lookup.rstrip('/')
    competitors = [col for col in df.columns if col == semrush_lookup or '/' in col or '.com' in col]

last_competitor = competitors[-1] if competitors else None

# --- 2. Dynamic Filter Preparation ---
# We now include ALL filters to give maximum flexibility
dropdown_options = [name for name, _ in targeted_filters]

# --- 3. Build the UI Widgets ---
filter_dropdown = widgets.Dropdown(
    options=dropdown_options,
    description='🎯 Filter:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

generate_button = widgets.Button(
    description='Generate Prompt Fu',
    button_style='primary',
    icon='magic',
    layout=widgets.Layout(width='200px')
)

output_canvas = widgets.Output()

# --- 4. The Action Callback ---
def on_generate_click(b):
    with output_canvas:
        clear_output(wait=True)
        selected_filter = filter_dropdown.value
        
        print(f"⏳ Generating hyper-customized expert prompt for '{selected_filter}'...\n")
        
        # Isolate the keywords
        keywords = next(kws for name, kws in targeted_filters if name == selected_filter)
        
        # Apply the Regex filter to the in-memory 'df'
        pattern = r'\b(?:' + '|'.join([re.escape(k) for k in keywords]) + r')\b'
        filtered_df = df[df["Keyword"].str.contains(pattern, case=False, na=False)].copy()
        
        if filtered_df.empty:
            print(f"⚠️ No keywords found matching the '{selected_filter}' criteria in this dataset.")
            return

        # Apply surgical scoring methodology
        scored_df = secretsauce.normalize_and_score_surgical(
            filtered_df, 
            semrush_lookup, 
            has_botify, 
            last_competitor, 
            reorder=False
        )

        scored_df.sort_values(by='Combined Score', ascending=False, inplace=True)
        top_df = scored_df.head(20)

        # Dynamically find the client column
        target_col = [col for col in top_df.columns if semrush_lookup.rstrip('/') in col][0]
        
        # Build the URL-aware and Botify-aware payload
        cols_for_llm = ['Keyword', 'Search Volume', target_col]
        
        if 'Client URL' in top_df.columns: cols_for_llm.append('Client URL')
        if last_competitor and last_competitor in top_df.columns: cols_for_llm.append(last_competitor)
        if 'Competitor URL' in top_df.columns: cols_for_llm.append('Competitor URL')
        
        # MAGIC: Add literal bleeding traffic data if Botify GSC data is connected
        missed_clicks_col = "No. of Missed Clicks excluding anonymized queries"
        if has_botify and missed_clicks_col in top_df.columns:
            cols_for_llm.append(missed_clicks_col)
            
        llm_context_df = top_df[cols_for_llm]
        csv_payload = llm_context_df.to_csv(index=False)

        # Construct the Upgraded Prompt Fu
        prompt_fu = f"""# ROLE AND GOAL
You are a seasoned, elite SEO strategist. I am providing you with a high-value keyword extraction for my client's domain ({semrush_lookup}). 

# CONTEXT
These keywords fall specifically under the "{selected_filter}" category. They have been mathematically scored for immediate ROI based on search volume, keyword difficulty, and our client's current ranking position. 

Crucially, this data includes URL-level intelligence: whether we currently have a ranking landing page ('Client URL'), our current position ('{target_col}'), our primary competitor's position ('{last_competitor}'), and their winning URL ('Competitor URL'). 
{"This data ALSO includes exact 'Missed Clicks' pulled directly from Google Search Console via Botify, showing literal traffic bleeding to competitors." if has_botify else ""}

# DATA
Here are the top {len(llm_context_df)} immediate opportunities in CSV format:

{csv_payload}
# INSTRUCTIONS
Please review this data and provide a brief, data-backed consulting soliloquy to present to the client. This is a zero-sum game, and we need to define our blocking and tackling strategy so the client doesn't spin their wheels. 

Walk me through:
1. **The Overarching User Intent:** What is the psychological driver behind these "{selected_filter}" queries? (e.g., Are they looking for local stores? Comparing reviews? Seeking definitive answers? Hunting for holiday deals?)
2. **A URL-Level "Blocking and Tackling" Strategy:** Analyze the provided URLs. Are we using the wrong page templates (e.g., Homepage vs. Blog vs. Local Store Page) compared to the winning competitor? Outline the specific page template strategy required for the "{selected_filter}" category.
3. **Top 5 Opportunities (The ROI Justification):** Categorize the top 5 targets into one of two buckets, justifying the ROI based on the competitor's hold on the SERP {"and the specific Missed Clicks volume" if has_botify else ""}:
   - "Striking Distance": We have an existing Client URL ranking (e.g., positions 11-40). Detail the specific on-page tweaks, internal linking, or link graph manipulation needed to close the gap.
   - "New Content Production": We have no ranking URL (blank) or are hopelessly outranked. Specify the net-new asset required to go head-to-head with the provided Competitor URL.
"""
        print(prompt_fu)

generate_button.on_click(on_generate_click)

display(widgets.HBox([filter_dropdown, generate_button]))
display(output_canvas)
```
